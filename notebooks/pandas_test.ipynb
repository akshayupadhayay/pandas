```python
# to select rows with index 1,2,5
employees.loc[[1,2,5]]
```

```python
# to select rows with index 1 to 5
employees.loc[1:5]
```

```python
# to select rows with index 5 to 10 and columns 'EMPLOYEE_ID', 'FIRST_NAME', 'LAST_NAME'
employees.loc[5:10, ['EMPLOYEE_ID', 'FIRST_NAME', 'LAST_NAME']]
```

```python
# Can change the index to a column name. This is useful when we want to select rows based on a column value.
employees.index = employees.EMPLOYEE_ID
employees.head()
```

```python
# to select rows with index 198 to 203 and columns 'FIRST_NAME', 'LAST_NAME'
employees.loc[198:203, ['FIRST_NAME', 'LAST_NAME']]
```

```python
employees.loc[198, 'FIRST_NAME'] = 'John'
employees.head()
```

```python
# set index to a column name and drop the original column. More pandas way to set index instead of assigning to index attribute. (employees.index = employees.EMPLOYEE_ID)
employees.set_index("EMPLOYEE_ID", inplace=True)
employees.head()
```

```python
# to reset the index to default integer index
employees.reset_index(drop=True, inplace=True)
employees.head()
```
