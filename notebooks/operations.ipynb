### Filtering Data

```python
bios = pd.read_csv('../data/external/bios.csv')
bios.tail()
```

```python
# to select rows where height_cm is greater than 210 and born_country is USA, and then select only the name, height_cm and athlete_id columns. 
# Finally, sort the result by height_cm in descending order.
result = bios[(bios['height_cm'] > 210) & (bios['born_country'] == 'USA')][['name', 'height_cm', 'athlete_id']]
result.sort_values(['height_cm'], ascending=[False])
result.head()
```

```python
# to select rows where the name column contains either 'Michael' or 'Rahul', ignoring case, and then sort the result by name in ascending order.
bios[bios['name'].str.contains('Michael|Rahul', case=False)].sort_values('name')
```

```python
# to select rows where the born_country column is either 'USA' or 'GBR', and the name column starts with 'Bill'.
bios[(bios['born_country'].isin(['USA','GBR'])) & (bios['name'].str.startswith('Bill'))]
```

```python
# to select rows where the born_country column is either 'USA' or 'GBR', and the name column starts with 'Bill' using query method.
bios.query('born_country in ["USA", "GBR"] and name.str.startswith("Bill")')
```

## Adding/Removing Columns

```python
people = pd.read_csv('../data/external/some-people.csv')
people.head()
```

```python
# to add a new column 'year' to the dataframe and set its value to 1985 for all rows.
people['year'] = 1985
people.head()
```

```python
import numpy as np
people['actual_year'] = np.where(people['born_country'] == 'USA', 2000, 2010)
people.head()
```

```python
# to drop the 'year' column from the dataframe. This will return a new dataframe without the 'year' column, but it will not modify the original dataframe unless we set inplace=True.
people.drop(columns=['year'])
people.head(4)
```

```python
people.head(2)
```

```python
# to add a new column 'age_after_5years' to the dataframe, which is the age of the person after 5 years. This is done by adding 5 to the 'age' column.
people['age_after_5years'] = people['age'] + 5
people.head(2)
```

```python
# to rename the column 'age_after_5years' to 'age_in_5_years'. This will modify the original dataframe because we set inplace=True.
people.rename(columns={'age_after_5years': 'age_in_5_years'}, inplace=True)
people.head(2)
```

```python
# to select rows where the first name is 'Albert'. We will first create a new column 'first_name' by splitting the 'name' column and taking the first part of the name. 
# Then we will filter the rows where 'first_name' is 'Albert'.
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split().str[0]
bios_new.head()
```

```python
# to select rows where the first name is 'Albert'.
bios_new.loc[bios_new['first_name'] == "Albert"]
```

```python
# to create a new column 'born_datetime' by converting the 'born_date' column to a datetime object. 
# Then we will extract the year from the 'born_datetime' column and store it in a new column 'born_year'.
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'], format='%Y-%m-%d')
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new.head()
```

#### Save dataframe to new csv file

```python
# to save the modified dataframe to a new CSV file called 'bios_new.csv' in the 'processed' folder. 
# We will set index=False to avoid saving the index as a separate column in the CSV file.
bios_new.to_csv('../data/processed/bios_new.csv', index=False)
```

```python
# to create a new column 'height_category' based on the values in the 'height_cm' column.
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x: 'Short' if x<165 else 'Average' if x<185 else 'Tall')
bios_new.head()
```

```python
# to create a new column 'weight_category' based on the values in the 'weight_kg' column. 
# We will use the apply method to apply a function that categorizes the weight into 'Lightweight', 'Middleweight' and 'Heavyweight'. 
# axis=1 is used to apply the function to each row of the dataframe. 
def categorize_height(row):
    if row['height_cm'] < 165 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] < 85:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
bios_new['weight_category'] = bios_new.apply(categorize_height, axis=1)
bios_new.head()
```

## Merging and Concatenating Data

```python
nocs = pd.read_csv('../data/external/noc_regions.csv')
bios = pd.read_csv('../data/external/bios.csv')
```

```python
# to merge the bios and nocs dataframes on the born_country and NOC columns respectively. 
# We will use a left join to keep all the rows from the bios dataframe and add the corresponding region information from the nocs dataframe.
bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
bios_new.head()
```

```python
# to select rows where the NOC_x column (which is the born_country from bios) is not equal to the region column (which is the region from nocs).
bios_new[bios_new['NOC_x'] != bios_new['region']][['name', 'born_country', 'NOC_x', 'region']]
```

```python
# to select rows where the born_country column is either 'USA' or 'GBR', 
# and then concatenate the resulting dataframes for USA and GBR into a new dataframe called new_df.
usa = bios_new[bios_new['born_country'] == 'USA']
gbr = bios_new[bios_new['born_country'] == 'GBR']

new_df = pd.concat([usa, gbr], ignore_index=True)
```

```python
# to display the first 2 rows of the new_df dataframe with USA and GBR athletes.
new_df.head(2)
```

```python
# to display the last 2 rows of the new_df dataframe with USA and GBR athletes.
new_df.tail(2)
```

```python
result = pd.read_csv('../data/external/results.csv')
combined_df = pd.merge(result, new_df, on='athlete_id', how='left')
combined_df.head()
```
